```python
INITIAL_VOL = df[df["timesteps"] == 0][DATA.VOLUME.value].iloc[0]
df[DATA.VOLUME.value] /= INITIAL_VOL
df.head()
```

Remove gamma-ramp timesteps

```python
df = pd.DataFrame.copy(df[df.timesteps >= 0.1])
df.timesteps -= 0.1
df.head()
```

remove unfinished simulations

```python
dfgrouped = df.groupby(["endo", "epi", "gamma"])
print("before:", df["timesteps"].count())
df_filtered = []
for k, group in dfgrouped:
  if group["timesteps"].count() <= 12:
    continue
  df_filtered.append(group)
df = pd.concat(df_filtered)
print("after:", df["timesteps"].count())
```

add pressure

```python
# df = pd.DataFrame.copy(raw_df)
IDEAL = str(datasetpath).lower().find("ideal") >= 0
print("using ideal:", IDEAL)

if IDEAL:
  max_pressure_computed = 20.0 # 20 for ideal, 24 for LV
  df["pressure"] = max_pressure_computed*df["timesteps"]
else:
  max_pressure_computed = 25.0 #16 for ideal, 24 for LV
  df["pressure"] = max_pressure_computed*df["timesteps"]
```

```python
df["pressure"].describe()
```

remove timesteps (no longer necessary)

```python
df.drop("timesteps", axis=1, inplace=True)
```

remove outliers

```python
dfgrouped = df.groupby(["endo", "epi", "gamma"])
print("before:", df["pressure"].count())
df_filtered = []
for k, group in dfgrouped:
  Q1 = group.quantile(0.03)
  Q3 = group.quantile(0.97)
  IQR = Q3 - Q1
  group = pd.DataFrame.copy(group[~((group < (Q1 - 3 * IQR)) |(group > (Q3 + 3 * IQR))).any(axis=1)])
  df_filtered.append(group)
df = pd.concat(df_filtered)
print("after:", df["pressure"].count())
```

```python
from scipy import stats
print("before:", df["pressure"].count())
df = pd.DataFrame.copy(df[(np.abs(stats.zscore(df)) < 3).all(axis=1)])
print("after:", df["pressure"].count())
```

```python
Q1 = df.quantile(0.01)
Q3 = df.quantile(0.99)
IQR = Q3 - Q1
print("before:", df["pressure"].count())
df = pd.DataFrame.copy(df[~((df < (Q1 - 3 * IQR)) |(df > (Q3 + 3 * IQR))).any(axis=1)])
print("after:", df["pressure"].count())
```

# Apply Interpolation in pressure-gamma

```python
dfgrouped = df.groupby(["endo", "epi"])
```

```python
max_pressure = np.round(df["pressure"].max())
max_pressure = min([25, max_pressure])
max_pressure
```

```python
df["pressure"].max()
```

```python
from scipy.interpolate import griddata, interp2d, RectBivariateSpline
from scipy.ndimage import gaussian_filter

nsteps = 50

xx = np.linspace(0, 20,nsteps+1)    # pressure -> 20
yy =  np.linspace(0, 0.30, nsteps+1) # gamma
X, Y = np.meshgrid(xx,yy)

new_grouped = dict()

for k, group in dfgrouped:
  new_grouped[k] = dict()
  new_grouped[k]["gamma"] = Y.ravel()
  new_grouped[k]["pressure"] = X.ravel()

  x = group["pressure"]
  y = group["gamma"]

  for col in group.columns:
    if col in ["gamma", "pressure", "endo", "epi"]:
      continue 
    z = group[col]
    first_interp = griddata((x,y), z, (X, Y), method="linear", rescale=True).ravel()
    nearest_interp = griddata((x,y), z, (X, Y), method="nearest", rescale=True).ravel()
    new_grouped[k][col] = np.where(np.isnan(first_interp), nearest_interp, first_interp)
```

```python
new_df_groups = []
for k, group in new_grouped.items():
  _gdf = pd.DataFrame(group)
  _gdf["endo"] = k[0]
  _gdf["epi"] = k[1]
  new_df_groups.append(_gdf)
df2 = pd.concat(new_df_groups)
```

```python
df = pd.DataFrame.copy(df2)
```

```python
df.dropna(inplace=True)
```

# Apply interpolation in endo-epi

```python
dfgrouped = df.groupby(["gamma", "pressure"])
```

```python
from scipy.interpolate import griddata, interp2d, RectBivariateSpline

nsteps = 20

xx = np.linspace(40, 80, nsteps+1) # endo
yy =  np.linspace(-80, -40,nsteps+1)  # epi
X, Y = np.meshgrid(xx,yy)

new_grouped = dict()

for k, group in dfgrouped:
  new_grouped[k] = dict()
  new_grouped[k]["endo"] = X.ravel()
  new_grouped[k]["epi"] = Y.ravel()

  x = group["endo"]
  y = group["epi"]

  for col in group.columns: 
    if col in ["gamma", "pressure", "endo", "epi"]:
      continue 
    z = group[col]
    first_interp = griddata((x,y), z, (X, Y), method="linear", rescale=True).ravel()
    nearest_interp = griddata((x,y), z, (X, Y), method="nearest", rescale=True).ravel()
    new_grouped[k][col] = np.where(np.isnan(first_interp), nearest_interp, first_interp)
```

```python
new_df_groups = []
for k, group in new_grouped.items():
  _gdf = pd.DataFrame(group)
  _gdf["gamma"] = k[0]
  _gdf["pressure"] = k[1]
  new_df_groups.append(_gdf)
df2 = pd.concat(new_df_groups)
```

```python
df = pd.DataFrame.copy(df2)
```

# Feature Engineering

Bins

```python
n_bins = 50

bins_keys = []
for v in df.columns:
  k = v+"-b"
  df[k] = (pd.qcut(df[v], q=n_bins, labels=False, duplicates='drop')).astype(int) +1
  bins_keys.append(k)
```

normalize pressure

```python
PRESSURE_NORM = 16.0
df["pressure"] /= PRESSURE_NORM
```

# Save important information in a json file

## Suggested Scale

```python
scales = dict()

for col in df.columns:
  
  scaled = False
  
  if col.find(DATA.VOLUME.value)>=0: # we already normalized volume
    scales[col] = 1.0 
  
  elif col.find("gamma")>=0:
    scales[col] = 1.0
  
  elif col.find("epi")>=0:
    scales[col] = 100.0
  
  elif col.find("endo")>=0:
    scales[col] = 100.0
  
  elif col.find("pressure")>=0:
    scales[col] = 1.0
  
  elif col.endswith("-b"):
    scales[col] = 1.0
  
  else:
    scales[col] = 100.0
```

## json

```python
import json

# save columns as 'headers' with indices of their col
headers = {col: i for i, col in enumerate(df.columns)}

# save min and max value
val_ranges = dict()
for col in df:
  val_ranges[col] = [float(df[col].min()), float(df[col].max())] 

norms = {
    "pressure": PRESSURE_NORM,
    "volume": INITIAL_VOL
}

to_save = dict(headers=headers, 
               ranges=val_ranges, 
               scales=scales, 
               n_classes=n_bins,
               norms=norms
               )

with open(datasetpath/"header.json", "w") as jsonfile:
  jsonfile.write(json.dumps(to_save))
```
